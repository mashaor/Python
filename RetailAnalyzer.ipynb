```python
class RetailAnalizer:
  def __init__(self):
    self.data_set = pd.read_csv("/content/retail_sales.csv")
    self.data_set.dropna(inplace=True)

  def total_sales_perproduct(self):
    return self.data_set.groupby('Product')['Sales'].sum()

  def best_selling_product(self):   
    return self.total_sales_perproduct().sort_values(ascending=False).index[0]

  def avarge_daily_sales(self):
    return self.data_set.groupby('Date')['Sales'].mean()

  def sales_over_time(self):
    self.data_set.groupby('Date')['Sales'].sum().plot(x='Date', y='Sales', kind='line')
    plt.show()

  def sales_per_product(self):
    self.data_set.groupby('Product')['Sales'].sum().plot(x='Product', y='Sales', kind='bar')
    plt.show()
```

```python
analyzer = RetailAnalizer()

print(analyzer.best_selling_product())
print(analyzer.avarge_daily_sales())
print(analyzer.sales_over_time())
print(analyzer.sales_per_product())
```
